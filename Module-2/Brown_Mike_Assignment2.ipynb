```python
NB_EPOCH = 20
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10 # number of outputs = number of digits
OPTIMIZER = SGD() # SGD optimizer, explained later in this chapter
N_HIDDEN = 128
VALIDATION_SPLIT = 0.2 # how much TRAIN is reserved for VALIDATION
```

---
### Spliting Data Into Train and Test Sets

```python
# data: shuffled and split between train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

---
### Shaping

```python
# X_train is 60000 rows of 28x28 values --> reshaped in 60000 x 784

RESHAPED = 784

X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
# normalize
X_train /= 255
X_test /= 255

print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# convert class vectors to binary class matrices

# Y_train = np_utils.to_categorical(y_train, NB_CLASSES) # From the book, code replaced below
Y_train = to_categorical(y_train, NB_CLASSES) # Updated for current version

#Y_test = np_utils.to_categorical(y_test, NB_CLASSES) # From the book, code replaced below
Y_test = to_categorical(y_test, NB_CLASSES) # Updated for current version
```

```python
NB_EPOCH = 20
BATCH_SIZE = 128
OPTIMIZER = SGD()

model = Sequential()
model.add(Input(shape=(RESHAPED,)))
model.add(Dense(N_HIDDEN, activation='relu'))
model.add(Dense(N_HIDDEN, activation='relu'))
model.add(Dense(NB_CLASSES, activation='softmax'))

model.compile(
    loss='categorical_crossentropy',
    optimizer=OPTIMIZER,
    metrics=['accuracy'],
    run_eagerly=True
)

history = model.fit(
    X_train,
    Y_train,
    batch_size=BATCH_SIZE,
    epochs=NB_EPOCH,
    verbose=VERBOSE,
    validation_split=VALIDATION_SPLIT
)

score = model.evaluate(X_test, Y_test, verbose=VERBOSE)
print("Test accuracy:", score[1])
```

## Accuracy Comparison and Explanation

In this assignment I trained a neural network with two hidden layers using the MNIST handwritten digits dataset. I tested different values for the number of epochs to see how additional training time would affect the accuracy of the model. The baseline model used 20 epochs, while the experiments tested both a smaller value (5 epochs) and a larger value (40 epochs). Changing the number of epochs allows the model to pass through the training data more or fewer times, which can impact how well the network learns the patterns in the data.

When the model was trained with only 5 epochs, the training accuracy reached about 90.3%, the validation accuracy was about 91.3%, and the test accuracy was approximately 91.17%. This lower accuracy suggests the model did not have enough time to fully learn the patterns in the handwritten digits. With the baseline configuration of 20 epochs, the model performed better. The training accuracy increased to about 94.39%, the validation accuracy reached about 94.77%, and the test accuracy improved to around 94.40%. This shows that giving the model more training cycles allowed it to adjust its weights more effectively and improve its predictions.

In the final experiment, I increased the number of epochs to 40. With this additional training time, the model achieved a training accuracy of about 96.37%, validation accuracy of about 96.21%, and a test accuracy of approximately 96.19%. This indicates that the model continued to improve as it saw the training data more times. Overall, increasing the number of epochs improved the accuracy across the training, validation, and test datasets because the model had more opportunities to reduce prediction errors. However, if the number of epochs becomes too large, the model could eventually begin to overfit the training data, where performance on new data may stop improving or begin to decrease.
